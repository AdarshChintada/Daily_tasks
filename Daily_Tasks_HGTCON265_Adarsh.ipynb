**Task of the day(27-09-23)**: For a random 2D numpy array build a histogram in python.

```python
import numpy as np
from matplotlib import pyplot as plt
marks_1 = np.array([100, 92, 95, 90, 85, 90])
marks_2 = np.array([99, 100, 94, 98, 92, 96])
result = np.vstack((marks_1, marks_2))
print(result)
plt.hist(result)
plt.show()




```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,10,1)
y = 2*x-5
plt.title('Line Graph')
plt.plot(x,y,color='red')
plt.show()
plt.xlabel('x-axis')
plt.ylabel('y-axix')
```

```python
#subplot
x = np.arange(1,10,1)
y1 = 2*x-5
y2 = 2*x+5
plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.subplot(1,2,2)
plt.plot(x,y2,color='g')
plt.xlabel('x-bar')
plt.ylabel('y-bar')
plt.show()
```

```python
#barplot

fruit = {"apple":30, "Banana":10, "Guava":15}
names = fruit.keys()
rupee = fruit.values()
colors = ['r','y','g']
plt.bar(names,rupee,color= colors)
plt.title("bar graph")
plt.xlabel('fruits')
plt.ylabel('cost')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Import libraries

# Step 2: Generate a random 2D NumPy array
random_2d_array = np.random.randn(100, 2)  # Replace (100, 2) with your desired shape

# Step 3: Flatten the array
flattened_array = random_2d_array.flatten()
print(flattened_array)

# Step 4: Create a histogram
plt.hist(flattened_array, bins=20, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histogram of Random 2D Array')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the histogram
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Custom months and average monthly temperatures for City X and City Y
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
city_x_temperatures = np.array([10, 12, 15, 20, 25, 30])
city_y_temperatures = np.array([8, 10, 14, 18, 22, 26])

# Create the 2D array for histogram plotting
hist= np.vstack((months,city_x_temperatures, city_y_temperatures))
plt.hist(hist)
plt.show()
```
