**Task of the day(27-09-23)**: For a random 2D numpy array build a histogram in python.

```python
import numpy as np
from matplotlib import pyplot as plt
marks = np.random.randint(10,100,size=(10,2))
flat = marks.flatten() #flatten method
print(flat)

x_ticks = np.arange(20,120,20)
plt.xticks(x_ticks)
plt.hist(flat, bins =5, label=["column1", "column2"])

plt.title("Histogram using 2d numpy array")
plt.xlabel("2d array")
plt.ylabel("Freqency")
plt.legend()



plt.show()




```

```python
import numpy as np
from matplotlib import pyplot as plt
marks = np.random.randint(10,100,size=(10,2))
flat = marks.flatten() #flatten method
print(flat)

x_ticks = np.arange(20,120,20)
plt.xticks(x_ticks)
plt.hist(flat, bins =5, histtype='step', label=["city1","city2"])

plt.title("Step type Histogram")
plt.xlabel("temperature")
plt.ylabel("frequency")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
random_2d_array = np.random.randint(2,100, size=(10,2))
flat = random_2d_array.flatten() #flatten method
print(flat)
plt.hist(flat, bins=20, edgecolor='black',label=["column1","column2"])
plt.title('Histogram of Random 2D Array')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
temp = np.array([
    [26, 33],
    [31, 23],
    [29, 31],
    [63, 31],
    [34, 36],
    [32, 40],
    [32, 40]
])
flat = temp.flatten() #flatten method
print(flat)
x_ticks = np.arange(10,60,5)
plt.xticks(x_ticks)
plt.hist(flat, bins = 5, label = ["city1","city2"])
plt.xlabel("temperature")
plt.ylabel("Frequency")
plt.title("City Temperature")
plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,10,1)
y = 2*x-5
plt.title('Line Graph')
plt.plot(x,y,color='red')
plt.show()
plt.xlabel('x-axis')
plt.ylabel('y-axix')
```

```python
#subplot
x = np.arange(1,10,1)
y1 = 2*x-5
y2 = 2*x+5
plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.subplot(1,2,2)
plt.plot(x,y2,color='g')
plt.xlabel('x-bar')
plt.ylabel('y-bar')
plt.show()
```

```python
#barplot

fruit = {"apple":30, "Banana":10, "Guava":15}
names = fruit.keys()
rupee = fruit.values()
colors = ['r','y','g']
plt.bar(names,rupee,color= colors)
plt.title("bar graph")
plt.xlabel('fruits')
plt.ylabel('cost')
plt.show()
```
